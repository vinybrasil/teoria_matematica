```
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```
A = np.array([[2,3],[2,1]])
```

```
autovalores = np.linalg.eigvals(A)
```

```
autovalores
```

```
autovetores = np.linalg.eig(A)
autovetores
```

```
eig_1 = autovetores[1][0]
eig_1
```

```
x_novo = np.arange(-20,20,0.1)
```

```
m = eig_1[1]/eig_1[0]
```

```
y_novo = []
for i in x_novo:
  y_novo.append(i*m)
```

```
#gerar a linha do autovetor
eig_2 = autovetores[1][1] #ERRADO
n = eig_2[1]/eig_2[0]
y_novo_2 = []
for i in x_novo:
  y_novo_2.append(i*n)
```

```
#funcao linear

def function(z, t):

    x, y = z

    dxdt = 2*x + 3*y
    dydt = 2*x + y
    
    return [dxdt, dydt] 

def y_1(x):       #onde a eq. é zero
  return (-2/3)*x
def y_2(x):
  return -2*x


y_zada = y_1(x_novo) #onde a equação é zero
y_zada_2 = y_2(x_novo)


x = np.linspace(-20, 20, 50)
y = np.linspace(-20, 20, 50)

X, Y = np.meshgrid(x, y)

NI, NJ = X.shape 
u, v = np.zeros(X.shape), np.zeros(Y.shape)
t = 0

for i in range(NI):
    for j in range(NJ):
        x = X[i, j]
        y = Y[i, j]
        yprime = function([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
fig= plt.figure(figsize=(20,12))
#Q = plt.quiver(X, Y, u, v, color='r')
u = u / np.sqrt(u**2 + v**2);
v = v / np.sqrt(u**2 + v**2);

#color = np.sqrt(((v-n)/2)**2 + ((u-n)/2)**2)
'''z = X*np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)
n = -2
color = np.sqrt(((dx-n)/2)**2 + ((dy-n)/2)**2) '''


Q = plt.quiver(X, Y,  u, v, color='red', scale=None, scale_units='inches')
plt.xlabel('$y_1$')
plt.ylabel('$y_2$')

'''for y20 in np.arange(0,15,1):
    #t = np.linspace(0, 60, 10)
    t = np.linspace(30, 80, 100)
    y0 = [random.randint(-20,20), random.randint(-20,20)]
    ys = odeint(function,y0, t)
    plt.plot(ys[:,0], ys[:,1], 'b-') # path
    plt.plot([ys[0,0]], [ys[0,1]], 'o') # start
    plt.plot([ys[-1,0]], [ys[-1,1]], 's') # end
'''
#y_zada = np.zeros(x_novo.shape)
#y_zada_2 = np.zeros(x_novo.shape)
plt.grid(c='black', linestyle='-', linewidth=2)
#plt.plot([-2], [-3], 'bo')


plt.plot(x_novo, y_novo, label='1º autovetor')        #plotando o primeiro autovetor
plt.plot(x_novo, y_novo_2, label='2º autovetor')      #plotando o segundo autovetor

plt.plot(x_novo, y_zada, label='dx/dt = 0')        #onde a eq é zero
plt.plot(x_novo, y_zada_2, label='dydx = 0')      #onde a eq é zero
plt.xlim([-20, 20])
plt.ylim([-20, 20])
plt.legend(loc='best', shadow=True, facecolor='white')
plt.savefig('phase-portrait.png')
```
