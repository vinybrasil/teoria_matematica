```
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
def function(z, t):

    x, y = z

    dydx = 4*x**2 + 9*y**2

    return [x, dydx] 
```

```
x = np.linspace(-80, 80, 100)
y = np.linspace(-60, 60, 100)

X, Y = np.meshgrid(x, y)

NI, NJ = X.shape 
u, v = np.zeros(X.shape), np.zeros(Y.shape)
t = 0
```

```
for i in range(NI):
    for j in range(NJ):
        x = X[i, j]
        y = Y[i, j]
        yprime = function([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
```

```
fig= plt.figure(figsize=(20,12))
Q = plt.quiver(X, Y, u, v, color='r')

plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
#plt.xlim([-2, 8])
#plt.ylim([-4, 4])
plt.savefig('phase-portrait.png')
```

```
def function(z, t):
    #orbita 50,80
    x, y = z
    a = 0.1
    b = 0.002
    c = 0.0025
    d = 0.2
    dxdt = -a*x +b*y*x
    dydt = -c*y*x +d*y
    
    return [dxdt, dydt] 


x = np.linspace(70, 85, 100)
y = np.linspace(30, 80, 100)

X, Y = np.meshgrid(x, y)

NI, NJ = X.shape 
u, v = np.zeros(X.shape), np.zeros(Y.shape)
t = 0

for i in range(NI):
    for j in range(NJ):
        x = X[i, j]
        y = Y[i, j]
        yprime = function([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
fig= plt.figure(figsize=(20,12))
u = u / np.sqrt(u**2 + v**2);
v = v / np.sqrt(u**2 + v**2);

Q = plt.quiver(X, Y,  u, v, color='red', scale=None, scale_units='inches')
plt.xlabel('$y_1$')
plt.ylabel('$y_2$')


#plt.xlim([-20, 20])
#plt.ylim([-20, 20])
plt.savefig('phase-portrait.png')
```
