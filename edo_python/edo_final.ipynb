```
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```
def equilibrio(x, t, mi, b1, b2, teta, psi, b0, m, y_n, i_ext, p_ext):
  s, p = x
  ds = ((psi*b1*s)/(teta+psi*b2)) + (p-psi*b1*p)/(teta+psi*b2) -i_ext + (psi*b0 + psi*b1*p_ext - m)/(teta + psi*b2);
  dp = ((mi*teta*b1*s)/(teta + psi*b2)) +((-mi*(teta*b1+b2)*p)/(teta+psi*b2)) - mi*y_n + (mi*teta*b1*p_ext + mi*b0 + mi*b2*m)/(teta + psi*b2);
  return ds, dp

#isolar e em função de p
```

```
def function(z, t):

    x, y = z

    dxdt = 2*x + 3*y
    dydt = 2*x + y
    
    return [dxdt, dydt] 

def y_1(x):       #onde a eq. é zero
  return -x
def y_2(x):
  return -2*x



x_novo = np.arange(-20,20,0.1)
y_zada = y_1(x_novo)
y_zada_2 = y_2(x_novo)


x = np.linspace(-20, 20, 50)
y = np.linspace(-20, 20, 50)

X, Y = np.meshgrid(x, y)

NI, NJ = X.shape 
u, v = np.zeros(X.shape), np.zeros(Y.shape)
t = 0

for i in range(NI):
    for j in range(NJ):
        x = X[i, j]
        y = Y[i, j]
        yprime = function([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
fig= plt.figure(figsize=(20,12))
#Q = plt.quiver(X, Y, u, v, color='r')
u = u / np.sqrt(u**2 + v**2);
v = v / np.sqrt(u**2 + v**2);

#color = np.sqrt(((v-n)/2)**2 + ((u-n)/2)**2)
'''z = X*np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)
n = -2
color = np.sqrt(((dx-n)/2)**2 + ((dy-n)/2)**2) '''


Q = plt.quiver(X, Y,  u, v, color='red', scale=None, scale_units='inches')
plt.xlabel('$y_1$')
plt.ylabel('$y_2$')

for y20 in np.arange(0,15,1):
    #t = np.linspace(0, 60, 10)
    t = np.linspace(30, 80, 100)
    y0 = [90, y20]
    ys = odeint(function,y0, t)
    plt.plot(ys[:,0], ys[:,1], 'b-') # path
    plt.plot([ys[0,0]], [ys[0,1]], 'o') # start
    plt.plot([ys[-1,0]], [ys[-1,1]], 's') # end


plt.plot(x_novo, y_zada)
plt.plot(x_novo, y_zada_2)
plt.xlim([-20, 20])
plt.ylim([-20, 20])
plt.savefig('phase-portrait.png')
```
